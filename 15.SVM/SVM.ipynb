# 1. For Forest Fires dataset

```python
# SVM Classification
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn import preprocessing
import seaborn as sns 

#from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
#from sklearn.preprocessing import StandardScaler
```

```python
forest=pd.read_csv(r"C:\Users\pawar\Downloads\forestfires (1).csv")
forest
```

```python
forest.describe()
```

```python
forest.isna().sum()
```

```python
#Complete forestfires dataset
label_encoder = preprocessing.LabelEncoder()
forest['month']= label_encoder.fit_transform(forest['month'])
forest['day']= label_encoder.fit_transform(forest['day'])
forest['size_category']= label_encoder.fit_transform(forest['size_category'])
forest
```

```python
'''array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
'''
X = forest.iloc[:, 0:30]
Y = forest.iloc[:, -1]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## Grid Search CV 

```python
clf = SVC()

parameters = {'kernel':['rbf'],
               'gamma':[100, 50, 5, 0.5, 0.1, 0.01, 0.0001],
               'C':[50, 15, 10, 6, 5, 0.1, 0.001, 0.0001]}

gsv = GridSearchCV(clf, param_grid = parameters, cv=10)

gsv.fit(X_train, y_train)
```

```python
gsv.best_params_
```

```python
gsv.best_score_
```

```python
model = SVC(C = 50, gamma = 0.0001)
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
