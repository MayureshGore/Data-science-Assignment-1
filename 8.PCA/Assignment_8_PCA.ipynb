# PCA implementation

```python
from sklearn.decomposition import PCA
```

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

df_pca=pca.fit_transform(norm)
df_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color='blue')
```

```python
# Final Dataframe
final_df=pd.concat([df['Type'],pd.DataFrame(df_pca[:,0:3],columns=['PC1','PC2','pc3'])],axis=1)
final_df
```

```python
import seaborn as sns
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

# 1.hierarchical Clustering

```python
# Import Libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
```

```python
# As we already have normalized data, create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
df3=df.copy()
df3['clustersid']=hclusters.labels_
df3
```

# 2.KMeans clustering

```python
wcss=[]
for i in range(1,8):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(norm)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,8),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
df4=df.copy()
df4['clusters3id']=clusters3.labels_
df4
```
